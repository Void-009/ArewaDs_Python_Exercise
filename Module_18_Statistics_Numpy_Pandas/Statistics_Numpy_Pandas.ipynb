# Question 5:
Count the number of rows and columns
* Filter the titles which contain python
* Filter the titles which contain JavaScript
* Explore the data and make sense of it

```python
import pandas as pd
import os

file_path = os.path.join(os.getcwd(), '../data/hacker_news.csv')

# Read the CSV file into a pandas DataFrame
try:
    # Read the data
    df = pd.read_csv(file_path)

    # 1. Count the number of rows and columns
    num_rows, num_columns = df.shape
    print(f"Number of rows: {num_rows}")
    print(f"Number of columns: {num_columns}")

    # 2. Filter the titles which contain the word 'python' (case insensitive)
    python_titles = df[df['title'].str.contains('python', case=False, na=False)]
    print(f"\nTitles containing 'python':")
    print(python_titles[['title']].head())  # Display the first few titles

    # 3. Filter the titles which contain the word 'JavaScript' (case insensitive)
    javascript_titles = df[df['title'].str.contains('javascript', case=False, na=False)]
    print(f"\nTitles containing 'JavaScript':")
    print(javascript_titles[['title']].head())  # Display the first few titles

    # 4. Explore the data
    print("\nData Overview:")
    print(df.head())  # Show the first few rows to understand the structure

    # Exploring basic stats and types
    print("\nData Info (types, non-null counts, etc.):")
    print(df.info())

    # Show basic statistics
    print("\nBasic Statistics (for numeric columns):")
    print(df.describe())

    # Check for missing values
    print("\nMissing Data Summary:")
    print(df.isnull().sum())

except FileNotFoundError:
    print(f"File not found at {file_path}. Please check the path.")
except pd.errors.EmptyDataError:
    print(f"The file {file_path} is empty.")
except pd.errors.ParserError:
    print(f"There was an issue parsing the file {file_path}.")
```
